## Attention
디코더에서 계산시 인코더의 모든 스텝에서 계산된 은닉 상태를 활용 가능.
은닉 상태마다 가중치를 다르게 적용하여 더 중요한 정보를 강조(attention)

●Attention 메커니즘
인코더의 모든 은닉 상태를 활용해서 디코더 계산
각각의 은닉 상태에 다른 가중치를 곱해 중요한 정보를 강조.
디코더가 입력 토큰마다 중요도를 다르게 부여
장점 : 정보 손실 줄이는데 효과적, 모든 타입스텝에서 생성된 은닉 상태를 참고하기 때문
단점 : 모든 은닉 상태 저장해야해서 연산량 증가, 그래서 인코더가 처리 가능한 타임스텝의 최대 개수를 지정해야하고 이로인해 입력 텍스트의 길이가 제한될 수 있음. 어텐션써도 한 번에 한 토큰씩 처리해야함.

●Transformer
어텐션 효과를 극대화하기 위해.
구글에서 2017년(attention is all you need)에서 발표
기존 인코더-디코더 구조를 유지하면서 순환 신경망을 제거 -> 입력 텍스트를 한 토큰씩이 아닌 한 번에 처리 가능.
인코더는 입력된 텍스트를 한 번에 모두 처리(타임 스텝 필요 없음), 어텐션 메커니즘을 써서 길이에 제한 있으나 빠름.
은닉 상태를 안쓰고 은닉 벡터,단어 벡터, 임베딩 벡터라는 표현을 사용.
디코더는 이전에 생성된 토큰을 참고하면서 새로운 토큰을 만듬.(자기회귀)
○self attention
기본 어텐션은 인코더 은닉 상태, 디코더 은닉 상태를 비교해 디코더가 특정 타입스텝에서 어떤 토큰에 집중해야하는지 학습,
Transforemr에서는 인코더에 입력되는 토큰만으로 어텐션 가중치를 학습(self attention)
1. 입력 텍스트의 각 토큰을 같은 밀집층에 통과시킴.
밀집층을 통과한 벡터 = 쿼리(query)
2. 같은 입력 텍스트를 두 번째 밀집층에 통과시킴 -> 키(key)
3. 쿼리 벡터, 키 벡터를 서로 곱해 어텐션 점수를 계산.(어텐션 행렬)
4. 입력 텍스트를 또 다른 밀집층에 통과시켜 값(value)벡터를 계산
5.어텐션 점수를 값 벡터에 곱해서 최종적인 셀프 어텐션을 출력.(각 입력 토큰이 다른 토큰들과 어느정도 관련있는지 반영한 은닉 벡터임)

쿼리 -> 현재 토큰이 다른 토큰에서 어떤 정보를 찾고 싶은지
key -> 각 토큰이 가지고 있는 특징 벡터
value ->  최종적으로 가중합되는 실제 정보 벡터
쿼리,키,벡터를 생성하는 밀집층의 가중치도 같이 학습됨.

○multi head attention
셀프 어텐션 연산을 수행하는 하나의 단위를 어텐션 헤드라고함. Transformer는 여러 개의 어텐션 헤드를 사용하는데 multi head attention이라고 함.
각 헤드에서는 쿼리, 키, 벡터를 생성하는 밀집층이 다르게 사용됨.
어텐션 헤드의 출력은 하나로 합쳐지고 밀집층 통과하여 어텐션 층의 최종 출력이 됨.

●층 정규화
입력 데이터를 정규화 해야하는데 딥러닝에선 여러 층을 거치면서 특성 스케일이 변할 수 있어서 단순 입력 정규화가 아닌 배치 정규화가 필요.(모든 특성을 평균 0, 분산 1로 정규화하면 정보 소실 가능)
○배치 정규화
보통 특성 맵의 크기는 (샘플 개수, 높이, 너비)
특정 채널의 데이터를 모아 평균, 분산 계산한 후 정규화를 적용(배치단위)(각 샘플의 채널별로 정규화)
장점 : 훈련 속도가 빨라지고 학습 과정이 안정화됨. but 텍스트에 적용하기는 어렵(샘플마다 길이가 달라서) -> 층 정규화
○층 정규화
각 샘플의 토큰마다 개별적으로 정규화 진행, 길이가 달라도 독립적으로 정규화 가능, 샘플 길이에 영향을 받지 않음

원본 Transformer 모델 -> 밀집층 후 드롭아웃 + 층 정규화
멀티 헤드 어텐션과 정규화 사이에는 잔차 연결(residual connection)(skip connection)이 추가.(ResNet에서 처음 도입)(층을 거진 출력에 입력값을 그대로 더함)
신경망 훈련시 가중치 업데이트가 뒤에서부터 진행되는데 잔차 연결이 추가되면 이 신호가 멀티 헤드 어텐션을 거치지 않고 직접 앞쪽 층으로 전달되어 신경망 층을 많이 쌓아도 효과적으로 훈련 가능.

○Feedforward Network
트랜스포머 인코더에서 멀티 헤드 어텐션과 층 정규화 다음에 나오는 밀집층을 말함.
보통 두 개의 밀집층으로 구성.
첫 번째 밀집층 : ReLU 활성화 함수 사용.
두 번째 밀집층 : 활성화 함수 사용 X
그 후 드롭아웃 적용 후 이 세 개의 층을 잔차연결이 감싼다.

○트랜스포머 인코더 블록
멀티 헤드 어텐션과 피드포워드 네트워크를 합침. -> 인코더 블록
MHA -> 정보 섞기(토큰 간 정보 교환)
FFN -> 표현력 증가(깊은 표현력)
입력 토큰은 단어 임베딩과 같은 벡터 표현으로 변환되는데 인코더 블록이 출력하는 은닉 벡터와 차원이 동일해서 동일한 인코더 블록을 여러 개 배치 가능.


○토큰 임베딩
-> 입력된 문자를 이해할 수 있도록 토큰을 숫자로 변환하는 방법
트랜스포머는 모든 토큰을 동시에 처리하기 때문에 토큰의 위치를 고려하지 않는 문제가 발생. 관계는 반영되나 위치는 고려되지 않음.
EX) i love you, you love me 에서 you의 어텐션 점수는 동일 but 단어는 위치에 따라 의미가 달라짐 -> 위치 정보가 필요

○위치 임베딩(인코딩)
사인 함수, 코사인 함수를 사용해서 토큰 위치에 따라 변하는 벡터 생성하고 단어(토큰) 임베딩에 더함.
ex) 임베딩 차원이 5고 10번째 토큰이 [0.1,0.2,0.6,0.3,0.5]로 이루어지면 각 원소 인덱스를 임베딩 벡터 전체길이(5)로 나눈후 이 값을 10,000의 거듭제곱한 값의 역수로 변환하고 10을 곱함.
그리고 짝수 원소는 사인 함수, 홀수는 코사인 함수를 적용하고 이 값을 원본 임베딩 벡터에 적용.->최종 단어 임베딩, 위치에 따라 값이 달라짐, 주기성을 가짐.
