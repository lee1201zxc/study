은닉층에는 활성화 함수를 사용해야함.
사용하지 않으면 선형적인 산술 계산만 진행하고 다음 층의 계산과 단순히 합쳐져서 앞 계산이 의미가 없어지기 때문

```python
import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled=train_input/255.0
train_scaled=train_scaled.reshape(-1,28*28)

train_scaled, val_scaled, train_target, val_target= train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

```python

inputs = keras.layers.Input(shape=(784,))
# 은닉층은 출력층의 뉴런보다는 많이 만들어야함.
dense1 = keras.layers.Dense(100, activation='sigmoid')
dense2 = keras.layers.Dense(10, activation='softmax')
model = keras.Sequential([inputs, dense1, dense2])
print()

# 이런 형태로도 가능
model= keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation='sigmoid', name='은닉층'),
    keras.layers.Dense(10, activation='softmax', name='출력층')
], name='패션 MNIST 모델')

# 이런 형태로도 가능! 세 가지 방법이 있음, 이게 가장 ㅎ편리한 듯?
model = keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(100, activation='sigmoid', name='hidden'))
model.add(keras.layers.Dense(30, activation='sigmoid', name='hiddens'))
model.add(keras.layers.Dense(10, activation='softmax', name='output'))

# 층에 대한 유용한 정보 얻을 수 있음
model.summary()

# 기본적으로 RMSprop경사 하강법 알고리즘을 사
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# 층 이름을 한글로하면 오류남, 여기서 batch_size도 변경 가능
model.fit(train_scaled, train_target, epochs=5)
```

##렐루(ReLU) 함수

시그모이드 함수는 값이 0~1로 압축되있어서 깊은 신경망에서 기울기 소실 문제 발생, 학습이 느려짐

max(0,z)임-> 음수에서 0으로 바꿈

###렐루 장점
계산이 빠르고 기울기 소실이 거의 없음(계속 음수만 받으면 0만 나와서 Dead ReLU문제 발생 -> Leaky ReLU사용)

이미지 처리에서 주로 사용

```python
model = keras.Sequential()
# 784가 아닌 원본 이미지 크기인 28,28로
model.add(keras.layers.Input(shape=(28,28)))
# Flatten -> 1차원으로 펼쳐줌
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

# Flatten에서 praram 0임
print(model.summary())

#데이터 다시 준비, reshape사용하지 않음
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
train_scaled=train_input/255.0

train_scaled, val_scaled, train_target, val_target= train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)

model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)

#sigmoid쓰고 3층으로 구성한 것보다 relu쓰고 2층구성이 더 좋음. 그만큼 relu가 좋다.
model.evaluate(val_scaled, val_target)
```

##1. SGD = 기본 경사 하강법 옵티마이저
모멘텀, 네스테로프 모멘텀을 EWMA라고도함(여러 번의 gradient 결과를 "부드럽게 평균"해서 사용)
###1.1 모멘텀
SGD클래스에서 모멘텀값을 0이상으로 설정(보통 0.9이상)하면 모멘텀 최적화고 이전의 그레디언트를 가속도 처럼 사용함(빠른 수렴)(방향 보정)<br><br>

###1.2 네스테로프 모멘텀
SGD클래스에서 nestrov값을 True로 바꾸면 네스테로프 가속 경사를 사용<br>
모멘텀 + 미래 gradient계산<br>
미래 위치에서 gradient를 계산함, 더 똑똑하고 빠르게 보정, 오버슈팅 방지

##2. 적응적 학습률 옵티마이저
모델이 최적점에 가까워질수록 학습률을 낮춰 안정적으로 수렴할 수 있게함.<br><br>


###2.1 Adagrad
각 파라미터마다 learning_rate를 다르게 조정

자주 업데이트되는 가중치는 천천히 움직이고 적게 움직이는 가중치는 빠르게 움직인다!

지금까지의 gradient 누적합을 사용 -> learning_rate가 작아져서 학습이 멈춤<br><br>

###2.2 RMSprop
Adagrad의 learning_rate 감소 문제 해결을 위해
최근 gradient만 계산함<br><br>
