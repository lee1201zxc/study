
Autograd: Automatic Differentiation
===================================

Central to all neural networks in PyTorch is the ``autograd`` package.
Let’s first briefly visit this, and we will then go to training our
first neural network.


The ``autograd`` package provides automatic differentiation for all operations
on Tensors. It is a define-by-run framework, which means that your backprop is
defined by how your code is run, and that every single iteration can be
different.

Let us see this in more simple terms with some examples.

autograd -> 자동 미분 제공

정적인 그래프가 아닌 실행하면서 바뀌는 그래프 -> Pytorch특징




Tensor
--------

``torch.Tensor`` is the central class of the package. If you set its attribute
``.requires_grad`` as ``True``, it starts to track all operations on it. When
you finish your computation you can call ``.backward()`` and have all the
gradients computed automatically. The gradient for this tensor will be
accumulated into ``.grad`` attribute.

To stop a tensor from tracking history, you can call ``.detach()`` to detach
it from the computation history, and to prevent future computation from being
tracked.

To prevent tracking history (and using memory), you can also wrap the code block
in ``with torch.no_grad():``. This can be particularly helpful when evaluating a
model because the model may have trainable parameters with
``requires_grad=True``, but for which we don't need the gradients.

There’s one more class which is very important for autograd
implementation - a ``Function``.

``Tensor`` and ``Function`` are interconnected and build up an acyclic
graph, that encodes a complete history of computation. Each tensor has
a ``.grad_fn`` attribute that references a ``Function`` that has created
the ``Tensor`` (except for Tensors created by the user - their
``grad_fn is None``).

If you want to compute the derivatives, you can call ``.backward()`` on
a ``Tensor``. If ``Tensor`` is a scalar (i.e. it holds a one element
data), you don’t need to specify any arguments to ``backward()``,
however if it has more elements, you need to specify a ``gradient``
argument that is a tensor of matching shape.


.requires_grad=True로 설정하면 그 텐서에 수행되는 모든 연산을 추적.



Create a tensor and set ``requires_grad=True`` to track computation with it



```python
'''
autograd -> 자동 미분 제공(gradient계산), 계산 과정 녹화
정적인 그래프가 아닌 실행하면서 바뀌는 그래프 -> Pytorch특징

requires_grad=True -> 그 텐서에 수행되는 모든 연산을 추적
계산 끝난 후, .backward()호출 시 모든 gradient가 자동 계산됨, gradient는 .grad에 저장됨
.detach()호출 시 추적 종료
with torch.no_grad(): 이 블록 안에서는 아예 연산 기록을 하지 않음, 모델 평가시 유용, 메모리 절약

Tensor + Function -> 계산 그래프 생성, 계산 순서?-> 자동 미분 시스템 핵심
.grad_fn -> 텐서가 어떤 연산으로 생성됐는지 가리키는 포인터, 역전파에 필수
.grad -> gradient 저장
텐서가 스칼라(원소 1개)면 .backward()에 인자가 필요 없음

예시 코드
x = torch.tensor(2.0, requires_grad=True)  # 추적 시작
y = x * x                                  # y = x^2
y.backward()                               # 미분 계산
print(x.grad)

'''
x = torch.ones(2, 2, requires_grad=True)
print(x)
```

Neural Networks
===============

Neural networks can be constructed using the ``torch.nn`` package.

Now that you had a glimpse of ``autograd``, ``nn`` depends on
``autograd`` to define models and differentiate them.
An ``nn.Module`` contains layers, and a method ``forward(input)``\ that
returns the ``output``.


A typical training procedure for a neural network is as follows:

- Define the neural network that has some learnable parameters (or
  weights)
- Iterate over a dataset of inputs
- Process input through the network
- Compute the loss (how far is the output from being correct)
- Propagate gradients back into the network’s parameters
- Update the weights of the network, typically using a simple update rule:
  ``weight = weight - learning_rate * gradient``

Define the network
------------------

Let’s define this network:


torch.nn -> 신경망 만드는 패키지, autograd에 의존함(모델은 nn이 만들고 미분 계산은 autograd가함)

nn.Module -> 여러 층을 포함하고 forward메소드 가지고 있으며 입력을 받아 출력값을 반환함

###학습 과정

1. 학습 가능한 가중치 가진 신경망 정의

2. 데이터셋을 반복해서 읽음

3. 입력을 모델에 통과시킴

4. 출력이 정답과 얼마나 다른지 계산

5. gradient를 네트워크 파라미터에 전파(역전파)

6. 가중치 업데이트


```python
import torch
# 신경망 레이어 모음
import torch.nn as nn
# 레이어 함수 버전(relu, softmax, cross_entropy같은거)
import torch.nn.functional as F

# 클래스 정의, nn.Module을 상속 받아야함.파라미터 자동 등록, GPU이용, autograd연결이 가능해짐
class Net(nn.Module):

    # 생성자
    # 모델 구조를 생성하고 파라미터 관리 시스템 작동
    # self -> 현재 생성 중인 객체 자신 -> net
    def __init__(self):
        # 부모 클래스 초기화, 부모는 nn.Module
        # 이거 안하면 파라미터 관리 시스템이 제대로 작동 안 함.(모델 작동 X)
        # super(현재 클래스, 현재 객체(net)), super ->상속 모듈(nn.Module), Net의 부모 클래스를 찾아라
        super(Net, self).__init__()
        # an affine operation: y = Wx + b
        # 선형 레이어(input 10차원, 출력 5차원)(밀집층임)
        self.linearss = nn.Linear(10, 5)

    # 입력이 모델 통과하는 경로 정의
    # x= 입력 데이터
    # output = net(input)이걸 쓰면 내부적으로 output = net.forward(input)이렇게됨
    def forward(self, x):
        # 생성자에서 만든 레이어 사용, 밀집층 통과시킴(5차원 나옴)
        x = self.linearss(x)
        # 레이어 통과한 결과 반환(모델 출력)(가중치와 bias값들)
        return x

# self -> 현재 생성 중인 객체 자신 -> net
# self.linear 에서 linear는 사용자가 임의 지정한 이름임
net = Net()
print(net)
```

You just have to define the ``forward`` function, and the ``backward``
function (where gradients are computed) is automatically defined for you
using ``autograd``.
You can use any of the Tensor operations in the ``forward`` function.

The learnable parameters of a model are returned by ``net.parameters()``



```python
# net.parameters() -> 모델 안의 weight과 bias반환
params = list(net.parameters())
print(params)
# weight크기
print(params[0].size())
# bias크기
print(params[1].size())
```

Let's try a random 1x10 input.
Note: expected input size of this net is 10.

```python
# 모델에 실제 데이터 넣어보기
input = torch.randn(1, 10)
# input이 모델에서 x에 해당함
print(input)
out = net(input)
# 밀집층 통과한 5개층
print(out)
```

Loss Function
-------------
A loss function takes the (output, target) pair of inputs, and computes a
value that estimates how far away the output is from the target.

There are several different
`loss functions <https://pytorch.org/docs/nn.html#loss-functions>`_ under the
nn package .
A simple loss is: ``nn.MSELoss`` which computes the mean-squared error
between the input and the target.

For example:

### 번역

파이토치는 mini-batch만 지원함


이제 lose계산, 역전파, 가중치 업데이트 과정을 거쳐야함


```python

output = net(input)
# target 랜덤 생성
target = torch.randn(5)  # a dummy target, for example
# numpy의 reshape랑 view랑 같다고 생각하셈
# target=torch.randn(1,5)로 해도 같음
target = target.view(1, -1)  # make it the same shape as output
# Loss 함수 지정
criterion = nn.MSELoss()
# Loss 계산
loss = criterion(output, target)
print(loss)
```

when we call ``loss.backward()``, the whole graph is differentiated
w.r.t. the loss, and all Tensors in the graph that has ``requires_grad=True``
will have their ``.grad`` Tensor accumulated with the gradient.

For illustration, let us follow a few steps backward:



```python
# loss를 만든 연산 정보
print(loss.grad_fn)  # MSELoss
# 이 연산 바로 이전에 연결된 연산들
# [0]첫번째연결노드, [0]실제 function 객체
print(loss.grad_fn.next_functions[0][0])  # Linear
```

Backprop
--------
To backpropagate the error all we have to do is to ``loss.backward()``.
You need to clear the existing gradients though, else gradients will be
accumulated to existing gradients.


Now we shall call ``loss.backward()``, and have a look at conv1's bias
gradients before and after the backward.



```python
# 모든 파라미터의 gradient를 0으로 초기화
net.zero_grad()     # zeroes the gradient buffers of all parameters

print('linearss.bias.grad before backward')
print(net.linearss.bias.grad)

# loss를 weight와 bias에 대해 미분(기울기 계산), 각 파라미터의 .grad에 저장
loss.backward()

print('linearss.bias.grad after backward')
# loss를 줄이기 위해 bias를 얼마나 바꿔야하는지 값 있음.
print(net.linearss.bias.grad)
```

Update the weights
------------------
The simplest update rule used in practice is the Stochastic Gradient
Descent (SGD):

     ``weight = weight - learning_rate * gradient``

We can implement this using simple Python code:

.. code:: python

    learning_rate = 0.01
    for f in net.parameters():
        f.data.sub_(f.grad.data * learning_rate)

However, as you use neural networks, you want to use various different
update rules.
To enable this, we built a small package: ``torch.optim`` that
implements all these methods. Using it is very simple:



```python
import torch.optim as optim
# 1스텝

# create your optimizer
# SGD방식으로  net.parameters를 업데이트, learning rate=0.01
optimizer = optim.SGD(net.parameters(), lr=0.01)

# in your training loop:
optimizer.zero_grad()   # zero the gradient buffers
output = net(input)
loss = criterion(output, target)
loss.backward()
# 학습, 가중치 업데이트
'''
for param in net.parameters():
    param = param - lr * param.grad
'''
optimizer.step()    # Does the update
```

      Observe how gradient buffers had to be manually set to zero using

It has the classes '0' to '9'.
The images in MNIST are of size 1x28x28, i.e. 1-channel images of 28x28 pixels in size.


Training an image classifier
----------------------------

We will do the following steps in order:

1. Load and normalizing the MNIST training and test datasets using
   ``torchvision``
2. Define a network
3. Define a loss function
4. Train the network on the training data
5. Test the network on the test data

## 1. Loading and normalizing MNIST

Using ``torchvision``, it’s extremely easy to load MNIST.



```python
import torch
from torchvision import datasets, transforms
```

The output of torchvision datasets are PILImage images of range [0, 1].
We transform them to Tensors of normalized range [-1, 1].
<div class="alert alert-info"><h4>Note</h4><p>If running on Windows and you get a BrokenPipeError, try setting
    the num_worker of torch.utils.data.DataLoader() to 0.</p></div>



```python
# Compose -> 여러 변환을 순서대로 적용하는 파이프라인
# ToTensor ->0~1범위로 스케일링
# Normalize -> mean,std값 정규화
transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
    ])

# transform -> 전처리 적용
trainset = datasets.MNIST('./data', train=True, download=True,
                   transform=transform)
testset = datasets.MNIST('./data', train=False,
                   transform=transform)
```

```python
# Training settings
batch_size = 64
test_batch_size = 1000
no_cuda = False
seed = 1
input_size = 28*28 # 784
num_classes = 10

use_cuda = not no_cuda and torch.cuda.is_available()

torch.manual_seed(seed)

device = torch.device("cuda" if use_cuda else "cpu")

# 데이터 준비를 병렬로해서 GPU가 쉬지 않게 함
# num_workers -> 0=메인 프로세스만 사용, 1이상=백그라운드에서 로딩
# pin_memory=True -> GPU학습시 메모리 복사 속도 향상
# 딕셔너리 형태임 그래서 뒤에 **
kwargs = {}
if use_cuda:
    kwargs.update({'num_workers': 1,
                   'pin_memory': True}
                 )

# DataLoader생성, 데이터를 배치 단위로 잘라서 제공하기 위함
# shffle=True -> 매 epoch마다 데이터 섞기
train_loader = torch.utils.data.DataLoader(dataset=trainset, batch_size=batch_size, shuffle=True, **kwargs)
test_loader = torch.utils.data.DataLoader(dataset=testset, batch_size=test_batch_size, shuffle=False, **kwargs)
```

## 2. Define a network

In this assignment, we will use logistic regression model

```python
net = nn.Linear(input_size, num_classes)

net = net.to(device)
```

## 3. Define a Loss function and optimizer
Let's use a cross entropy loss and SGD with momentum.


```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
# momentnum -> 이전 방향 기억해서 관성처럼 이동
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

## 4. Train the network

This is when things start to get interesting.
We simply have to loop over our data iterator, and feed the inputs to the
network and optimize.



```python
# training settings
epochs = 10
# 몇 스텝마다 로그를 출력할지
log_interval = 100
save_model = True
```

```python
def train_epoch(model, device, train_loader, optimizer, criterion, epoch):
    # 모델을 학습 모드로 변경(Dropout활성화, BatchNorm학습 모드 적용)
    model.train()
    # batch_idx->현재 몇 번째 배치인지
    for batch_idx, (data, target) in enumerate(train_loader):
        #데이터를 올바른 디바이스(GPU)로 이동(둘이 같아야 함)
        data, target = data.to(device), target.to(device)
        #Flatten, data.size(0)-> batch크기
        #ex) (64,1,28,28) -> (64,784)
        data = data.view(data.size(0), -1)
        #그래디언트 초기화
        optimizer.zero_grad()
        #순전파, 계산 그래프 기록
        output = model(data)
        #손실계산하기
        loss = criterion(output, target)
        #역전파 과정
        loss.backward()
        #가중치 업데이트
        optimizer.step()
        #로그 출력
        if batch_idx % log_interval == 0:
            print(f'Train Epoch: {epoch} [{batch_idx * len(data)}/{len(train_loader.dataset)} '
                  f'({100. * batch_idx / len(train_loader):.0f}%)]\tLoss: {loss.item():.6f}')
```

```python
def test(model, device, test_loader, criterion):
    # 모델을 평가모드로 전환
    model.eval()
    # 전체 손실 합계 저장
    test_loss = 0.0
    # 맞춘 개수 카운트
    correct = 0
    # gradient 계산X->메모리 절약
    with torch.no_grad():
        for data, target in test_loader:
            # 디바이스 이동
            data, target = data.to(device), target.to(device)
            # Flatten
            data = data.view(data.size(0), -1)
            #예측 = model.forward(data), 모델에 데이터 넣기,
            #output 값 -> 각 클래스 별 점수
            output = model(data)
            #배치 손실 누적
            #criterion(output, target)->현재 배치의 평균 loss
            #.itme()-> loss는tensor인데 이걸 그냥 숫자로 변환
            #data.size(0)->배치 크기(64), 나중에 전체 평균 구할때 배치 크기가 다르면 잘 못 구해짐(배치 사이즈 크든 작든 같은 영향임)
            test_loss += criterion(output, target).item() * data.size(0)
            #예측 라벨 뽑기
            #argnax->가장 큰 값의 인덱스 반환(가장 확률(점수) 높은 클래스), 각각의 데이터마다 여러개 pred=길이 64짜리 벡터
            #dim=1->가로방향(클래스 방향), keepdim=True->기본적으로 argmax는 차원을 줄이는데 target과 shape맞추기 위해씀
            pred = output.argmax(dim=1, keepdim=True)
            #맞은 개수 세기
            '''
            pred.shape    # (64, 1)
            target.shape  # (64,)
            둘이 달라서 맞춰야함 target.view_as(pred)로
            pred.eq()->각 원소 비교해서 같으면 True
            sum->이 배치에서 맞은 개수 총합
            correct에는 지금까지 돌면서 모든 데이터 맞은 개수가 들어감(한 배치만이 아닌)
            '''
            correct += pred.eq(target.view_as(pred)).sum().item()
    #전체 데이터 기준 평균 손실(앞에서 모든 샘플의 손실 구해놔서 전체 데이터 수로 나눔)
    test_loss /= len(test_loader.dataset)
    # loss,정확도 출력
    print(f'\nTest set: Average loss: {test_loss:.4f}, '
          f'Accuracy: {correct}/{len(test_loader.dataset)} '
          f'({100. * correct / len(test_loader.dataset):.2f}%)\n')
```

```python
for epoch in range(1, epochs + 1):
    train_epoch(net, device, train_loader, optimizer, criterion, epoch)

test(net, device, test_loader, criterion)
```
